```python
# count number of likes


data = data_original

data.head()

from google.colab import files
data.to_csv('dweeds.csv', encoding = 'utf-8-sig') 
files.download('dweeds.csv')
```

```python
days_and_likes = {}
row_num = 0
for day in set(data['Date_new']):
  tweets_on_day = data.loc[data['Date_new'] == day]
  print(day)
  for idx, tweet in tweets_on_day.iterrows():
    if day not in days_and_likes.keys():
      days_and_likes[day] = 0
    likes_on_tweet = tweet['# of likes']
    days_and_likes[day] += int(likes_on_tweet)
    row_num += 1
    # print("row num = " + str(row_num))

print(days_and_likes)
```
